# What is the time series
- Time series is a series of data points collected over an interval of time, where each point represents data at a specific timestamp. Python is one programming language used to help conduct time series analysis.

- **Overview of Time Series Analysis Techniques**

- Time series analysis encompasses several techniques that help in understanding and forecasting data.

1. **Date/Time Indexing** : Indexing data by date and time is the first step in time series analysis.This allows for time-based operations and analyses.
- **Importance** : Proper Indexing ensures that data is accurately aligned in time,which is essential for meaningful analysis and forecasting.

2. **Resampling**
   - **Downsampling** : Reducing the frequency of data points by aggregating them over a specific period.Example: Converting daily data to monthly data by averaging.

   - **Upsampling** : Incresing the frequency of data point by interpolation values.Exmple: Converting monthly data to daily data using interpolation techniques.
   
   - **Use Cases** : Adjusting the granularity of data for different analysis need.


### Converting Daily Data to Monthly Data by Averaging

In this section, we will downsample daily data to monthly data by calculating the average for each month.

#### Example:

Suppose we have daily temperature data for the month of January:

| Date       | Temperature (°C) |
|------------|------------------|
| 2024-01-01 | 10               |
| 2024-01-02 | 12               |
| 2024-01-03 | 11               |
| 2024-01-04 | 13               |
| 2024-01-05 | 14               |
| ...        | ...              |
| 2024-01-30 | 15               |
| 2024-01-31 | 16               |

To convert this daily data into monthly data by averaging, we calculate the average temperature for all the days in January.

#### Calculation:

\[
\text{Average Temperature for January} = \frac{10 + 12 + 11 + 13 + 14 + \ldots + 15 + 16}{31}
\]

Assuming the total sum of temperatures for January is 400°C:

\[
\text{Average Temperature for January} = \frac{400}{31} \approx 12.9°C
\]

So, the monthly data point for January would be **12.9°C**.

This technique helps in reducing data granularity and summarizing trends over longer periods.

#### Calculating the 3-Day Rolling Average

To compute a 3-day rolling average, we calculate the average number of steps over each 3-day window:

- **For August 03**: 
  \[
  \text{Average} = \frac{4000 + 5000 + 3000}{3} = 4000 \text{ steps}
  \]

- **For August 04**: 
  \[
  \text{Average} = \frac{5000 + 3000 + 6000}{3} = 4667 \text{ steps}
  \]

- **For August 05**: 
  \[
  \text{Average} = \frac{3000 + 6000 + 7000}{3} = 5333 \text{ steps}
  \]

- **For August 06**: 
  \[
  \text{Average} = \frac{6000 + 7000 + 2000}{3} = 5000 \text{ steps}
  \]

- **For August 07**: 
  \[
  \text{Average} = \frac{7000 + 2000 + 10000}{3} = 5000 \text{ steps}
  \]

**Note**: For the first two days (August 01 and August 02), there isn’t enough data to compute a 3-day rolling average, so those values are not calculated.

# Resampling
- Resampling is a powerful technique in time series analysis that involves changing the frequency fo time series data.This technique is essential for summarizing data,detection trends,and aligining data to common time intervals.

**what is Resampling**
- Resmapling is the process of altering the frequency of timee series data.It is used to convert data from one time frequency to another,which is particular useful for summarizing large dataset,smooth data,and aligning datasets with differnt frequencies
- **Downsampling**: Reducing the frequency of the data(e.g. from daily to month)
- **upsampling** : Increasing the frequency of the data(e.g. from monthly to daily)

```python
import pandas as pd
#sample data creation
data=pd.date_range(start='2023-01-01',periods=365,freq='D')
sales=np.random.randint(50,200,size=365)
df=pd.DataFrame({'date':data,'sales':sales})
print(df.head())
```

**Explanation** 

- The example create a DataFrame with daily sales data for one year.
- This dataset can then be resampled to a different frequency for analysis.

**Downsampling Aggregating Data** :
- Downsampling reduces the frequency of data by aggregating it over a specific period.This technique is useful for summarizing data,smoothing out short-term fluctuations,and identifying long term trends.

```python
#convert 'date' to datetime and set it index


df['date']=pd.to_datetime(df['date'])
df.set_index('date',inplace=True)

#Downsample to monthly frequency and aggregate using mean
monthly_sales=df.resample('M').mean()
print(monthly_sales)
```

**Explanation**

- The `resample()` function method changes the frequency to monthly.
- The `mean()` function calculate the average sales for each month.
- The aggregation helps in summarizing the data to show monthly trends.

# Upsampling: Increasing Frequency
- Updampling increases the frequency of time series data by creating additional time points and interpolating the missing values.This technique is useful for creating a more detailed view of the data and preparing it for more granular analysis.

```python
#Upsample to daily frequency and interpolate missing values.
daily_sales=monthly_sales.resample('D').interpolate(method='linear')
print(daily_sales)
```

**Explanation**

- The `resample('D')` method increases the frequency to daily.
- The `interpolate(method='linear') function fills in the missing values using linear interpolation.
- This provides a finer granularity of data,which can be useful  for more detailed analysis.

# Rolling Windows
- Rolling windows or moving windows are an analytical technique used in time series data to compute statistical measures over a defined susbset of data points.This techniques instrumental in understanding trends,smoothing data and evaluating variability by applying a function to a rolling window of data points

**Rolling Windows Concept**

- A rolling window invloves applying a statistical function to a subset of data points that moves across the entire dataset.Each subset is defined by the window size,which dictates how many data points are include in each calculation.
- **Purpose** : The primary goals is to analyze the data within a specified window to smooth out short term fluctuations  and higlight long term trends and patterns.Rolling windows are used to generate metrics that capture the behaviour of the data over time.

**Example**: Imagine analyzing a time series of daily temperatures. By applying a rolling window of 7 days, you can calculate the average temperature over each 7-day period, which helps in smoothing out daily fluctuations and revealing longer-term temperature tren

## Calculating Rolling Mean

**Rolling Mean**

- The rolling mean,or moving average,calculate the average of data points within a rolling window.This helps to smooth out short term variations and highlight longer term trends.
- **Applications** : By averaging the values within each window,the rolling mean provides a smoothed version of the time series,making it easier to see underlying patterns.

```python
#calculating the rolling mean
import pandas as pd
import numpy as np
#sample data:Daily sales over a year
data=pd.date_range(start='2023-01-01',periods=365,freq='D')
sales=np.random.randint(50,200,size=365)
df=pd.DataFrame({'date':data,'sales':sales})
df.set_index('date',inplace=True)

#calculating rolling mean with a 7-day window
df['rolling_mean']=df['sales'].rolling(window=7).mean()
print(df.head(10))
```

**Explanation**:

- **Data Creation** : The DataFrame `df` contains daily sales data for one year.
- **Rolling window** : `rolling(window=7) creates a rolling window of 7 days.
- **Mean Calculations** : `mean()` computes the average sales for each 7-day window.
- **Result** : The `rolling_mean` column shows a smoothed version of sales data,highlighting longer term trends while reducing daily noise.

**Calculating Rolling Standard Deviations**

- The rolling standard deviations measures the variability or spread the data within each rolling window.It provides insights into the consistency or volatility of the time series.
- **Applications** : By calculating the standard deviations within each window,we can assess how much the data varies over different periods.

```python
# calculating rolling standard deviations
df['rolling_std']=df['sales'].rolling(window=7).std()
print(df.head(10))
```

**Explanation**

- **Rolling Window** : `rolling(window=7)defines a 7-days rolling window.
- **Standard Deviations calculations** : `std()` computes the standard deviation of sales with each window.
- **Result** : The `rolling_std` column show how sales variability changes over time,indicating periods of higher or lower volatility.

**Applying custom Functions with Rolling Windows**

**Custom Functions**

- We can apply custom functions to rolling windows to perform specialized analyses,such as calculating the range(difference between maximum and minimum values)within each window.
- **Applications** : Custom functions allow to tailor the analysis to specific needs,providing more detailed insights based on our requirements.

```python
#Define a custom function to calculate the range within each window
def range_function(series):
    return series.max() - series.min()

#Apply the custom function with a 7-days window
df['rolling_range']=df['sales'].rolling(window=7).apply(range_function)
print(df.head(10))
```

**Explanation**
- **Custom function** : `range_function(series)` calculates the ranges(maximum-minimum) within each window.
- **Rolling Application** : `apply(range_function)` applies this function to each 7-day window.
- **Result** : The `rolling_range` column provides insights into the variability and extremes in sales data over time.

**Visualizing Rolling Statistics**

**Visualization**

- Visualizing rolling statistics helps to udnerstand trends,patterns,and the impact of smoothing on time series data.
- **Application** : Plotting the origional data alongside rolling stastistics provide a graphical representation of how rolling metrics evolve over time.


```python
import matplotlib.pyplot as plt
#plot origional sales data and rolling stastics

plt.figure(figsize=(12,6))
plt.plot(df.index,df['sales'],label='Original Sales',color='blue',alpha=0.5)
plt.plot(df.index,df['rolling_mean'],label='Rolling Mean(7 days)',color='orange')
plt.plot(df.index,df['rolling_std'],label='Rolling std dev(7 days)',color='green')
plt.title('Rolling statistics of Daily sales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

**Explanation:**

- **Plotting:**  plt.plot() creates plots for the original sales data, rolling mean, and rolling standard deviation.

- **Result:** The plot highlights how rolling statistics reveal trends and variability, making it easier to analyze the data’s behavior over time.

# Calculating Trend, Seasonal, and Residual Components in Time Series Analysis For 

In time series analysis, the observed time series (\(Y_t\)) is typically decomposed into three main components:
1. **Trend (\(T_t\))**
2. **Seasonal (\(S_t\))**
3. **Residuals (\(R_t\))**

## 1. Trend Component (\(T_t\))

The trend component represents the long-term movement or direction in the time series data.

### Calculation:
The trend can be calculated using various methods such as moving averages, polynomial regression, or exponential smoothing. 

#### Example:
Assume we use a simple moving average with a window of 3 months. For March, the trend component is the average of the time series values for January, February, and March.

\[
T_{\text{March}} = \frac{Y_{\text{January}} + Y_{\text{February}} + Y_{\text{March}}}{3}
\]

## 2. Seasonal Component (\(S_t\))

The seasonal component captures the repeating patterns or cycles in the data, such as monthly effects that repeat each year.

### Formula:
To calculate the seasonal component (\(S_t\)) for a particular time \(t\), you subtract the trend component (\(T_t\)) and the residual component (\(R_t\)) from the observed time series value (\(Y_t\)):

\[
S_t = Y_t - T_t - R_t
\]

If the residuals (\(R_t\)) are not available, you can approximate the seasonal component by:

\[
S_t \approx Y_t - T_t
\]

#### Example:
Assume we have the following data for a particular month:
- Observed time series value (\(Y_t\)): 175
- Trend component (\(T_t\)): 155

To approximate the seasonal component (\(S_t\)):

\[
S_t \approx Y_t - T_t = 175 - 155 = 20
\]

## 3. Residual Component (\(R_t\))

The residual component represents the random noise or irregular variation not captured by the trend or seasonal components.

### Formula:
To calculate the residual component (\(R_t\)), subtract the sum of the trend and seasonal components from the observed time series value (\(Y_t\)):

\[
R_t = Y_t - (T_t + S_t)
\]

#### Example:
Using the same data:
- Observed time series value (\(Y_t\)): 175
- Trend component (\(T_t\)): 155
- Seasonal component (\(S_t\)): 20

To calculate the residual:

\[
R_t = Y_t - (T_t + S_t) = 175 - (155 + 20) = 175 - 175 = 0
\]

1. **Trend Component** (\(T_t\)):
   - Calculated using methods such as moving averages or polynomial regression.

2. **Seasonal Component** (\(S_t\)):
   - Approximate if residuals are not available:
     \[
     S_t \approx Y_t - T_t
