# Environmental Data Analysis

Visualize environmental data to analyze trends in air quality over time. Use a dataset of air quality measurements (e.g., PM2.5 levels) from different locations and include:

- Line plots to show daily average PM2.5 levels over a year.
- Heatmaps to visualize PM2.5 levels across different locations and months.
- Scatter plots to show the correlation between PM2.5 levels and weather conditions (e.g., temperature, humidity).

__Hint:__ Use `plt.plot()` for line plots, `sns.heatmap()` from Seaborn for heatmaps, and `plt.scatter()` for scatter plots.

__Expectation:__ Visualizations depicting trends in air quality, spatial and temporal variations, and correlations with weather conditions.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load datasets
data=pd.read_csv(r"C:\Users\pc\Downloads\PRSA_data_2010.1.1-2014.12.31.csv")
data
```

```python
# Clean the data
data=data.drop_duplicates()
```

```python
# line plot to show daily average PM2.5 level over a year.
# Assuming the data is ordered by date,we can plot the PM2.5 levels directly
plt.figure(figsize=(10,6))
plt.plot(data['pm2.5'])
plt.title('Daily PM2.5 Levels')
plt.xlabel('Time(Days)')
plt.ylabel('PM2.5 Levels(Âµg/mÂ³)')
plt.show()
```

```python
# Heatmaps to visualize PM2.5 levels across different locations and month.
# Aggregates the data to get the mean PM2.5 levels for each locations and month.
data_clean = data.dropna(subset=['pm2.5', 'cbwd', 'month', 'year'])
pivot_table = data_clean.groupby(['cbwd', 'month'])['pm2.5'].mean().unstack()

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap='YlGnBu', linewidths=.5)
plt.title('Average PM2.5 Levels Across Different Locations and Months')
plt.xlabel('Month')
plt.ylabel('Location (cbwd)')
plt.show()
```

```python
# Scatter plots to show the correlation between PM2.5 levels and weather conditions (ex: Temp).
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TEMP', y='pm2.5', data=data, alpha=0.5)
plt.title('Correlation between PM2.5 Levels and Temperature')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('PM2.5 Levels (Âµg/mÂ³)')
plt.grid(True)
plt.show()
```
