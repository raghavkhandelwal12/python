**Project: Data Science Job Salaries Analysis**
**Dataset**

Data on job salaries for data science positions across different years. A suitable dataset is the Data Science Job Salaries Dataset from Kaggle.

**Visualization Goals**

- Analyze trends in data science salaries over different years.
- Compare salary distributions across various job titles.
- Visualize the impact of experience and location on salaries.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load the dataset
salary_data = pd.read_csv(r"C:\Users\pc\Downloads\data_science_salaries.csv")
```

```python
# Convert the 'year' column to datetime for better handling
salary_data['work_year'] = pd.to_datetime(salary_data['work_year'], format='%Y')
```

```python
# Convert the 'year' column to datetime for better handling
salary_data['work_year'] = pd.to_datetime(salary_data['work_year'], format='%Y')
```

```python
# Line plot for salaries over the years
plt.figure(figsize=(12, 6))
sns.lineplot(data=salary_data, x='work_year', y='salary', hue='job_title', palette='Set1')
plt.title('Data Science Job Salaries Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Salary (USD)')
plt.legend(title='Job Title')
plt.grid(True)
plt.show()
plt.figure(figsize=(12, 6))
sns.boxplot(data=salary_data, x='job_title', y='salary', color=sns.color_palette('Set2')[0])
plt.title('Salary Distribution by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Salary (USD)')
plt.xticks(rotation=45)
plt.show()

# FacetGrid for salary trends by experience level
salary_data['experience_level'] = pd.Categorical(salary_data['experience_level'], 
                                                  categories=['Junior', 'Mid-level', 'Senior', 'Lead'], 
                                                  ordered=True)
g = sns.FacetGrid(salary_data, col="experience_level", hue="job_title", height=4, aspect=1)
g.map(sns.lineplot, "work_year", "salary")
g.add_legend()
plt.title('Salary Trends by Experience Level')
plt.xlabel('Year')
plt.ylabel('Salary (USD)')
plt.show()
```
