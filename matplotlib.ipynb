# What is Matplotlib?
- Matplotlib is a low level graph plotting library in python that serves as a visualization utility.
- Matplotlib was created by `john D.Hunter`.
- Matplotlib is open source and we can use it freely.



```python
# checking matplotlib version
import matplotlib
print(matplotlib.__version__)
```

# Matplotlib Pyplot
**Pyplot**
- Most of the matplotlib utilities lies under the `pyplot` submodule ,and are usually imported under the plt alias.

```python
# draw a line in a diagram from position (0,0) to position (6,250)
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([0,6])
ypoints=np.array([0,250])
plt.plot(xpoints,ypoints)
plt.show()
```

# Matplotlib Plotting
**Plotting x and y points**
- The `plot()` function is used to draw point(markers) in a diagram.
- By default the `plot()` function draws a line from point to point.
- The function takes the specifying parameter for specifying points in the diagram.
- Parameter 1 is an array containing the points on the `x-axis`.
- Parameter 2 is an array containing the points on the `y-axis`.

```python
# draw a line in a diagram from position (1,3) to postion(8,10)
import matplotlib.pyplot as plt
import numpy as np
xpoint=np.array([1,8]) #this is the horizontal axis
ypoint=np.array([3,10]) #this is the vertical axis
plt.plot(xpoint,ypoint)
plt.show()
```

# Plotting without line
- To plot only the markers ,we can use shortcut string notation parameter `o` which means `rings`.

```python
import matplotlib.pyplot as plt
import numpy as np
xpoint=np.array([1,8])
ypoint=np.array([3,10])
plt.plot(xpoint,ypoint,'o')
plt.show()
```

# Multiple points
- We can plot as many points like,just make sure we have the same number of points in both axis.

```python
# Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])
plt.plot(xpoints,ypoints)
plt.show()
```

# Marker size
- We can use the keyword argument `markersize` or the shorter version ,`ms` to set the size of the markers.

```python
# set the size of the markers to 20
import matplotlib.pyplot as plt
import numpy as np
ypoint=np.array([3,8,1,10])
plt.plot(ypoint,marker='o',ms=20)
plt.show()
```

# Marker color
- We can use the keyword argument `markeredgecolor` or the shorter `mec` to set the color of the edge of the markers.


```python
# set the edge color of the red
import matplotlib.pyplot as plt
import numpy as np
ypoint=np.array([3,8,2,10])
plt.plot(ypoint,marker='o',mec='r',ms=30)
plt.show()
```

- We can use the keyword argument `markerfacecolro` or the shorter `mfc` to set the color inside the edge of the markers.


```python
import matplotlib.pyplot as plt
import numpy as np
ypoint=np.array([3,8,1,10])
plt.plot(ypoint,marker='o',ms=20,mfc='r')
plt.show()
```

# Multiple Lines
- We can plot as many lines as we like by simply adding more `plt.plot()` function

# Matplotlib Labels and Title
**create Labels for a Plot**
- With Pyplot we can use the `xlabel()` and `ylabel()` functions to set a label for the x-axis and y-axis

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x,y)
plt.xlabel('Average Pulse')
plt.ylabel('calories burnage')
plt.show()
```

# Create a Title for a Plot
- With Pyplot ,we can use the `title()` function to set title for the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x,y)
plt.title('sports watch data')
plt.xlabel('average pulse')
plt.ylabel('calories burnage')
plt.show()
```

# Set Font Properties for Title and Labels
- We can use the `fontdict` parameter in `xlabel()`,`ylabel()`,and `title()` to set font properties for the title and labels.

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}
plt.title('Sports Watch Data',fontdict=font1)
plt.xlabel('Averge Pulse',fontdict=font2)
plt.ylabel('calories burnage',fontdict=font2)
plt.plot(x,y)
plt.show()
```

# Position the Title
- We can use the `loc` parameter in `title()` to position the title.
- Legal values are 'left','right' and 'center'.Default values is center

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.title('sports watch data',loc='left')
plt.xlabel('average pulse')
plt.ylabel('calories burnage')
plt.plot(x,y)
plt.show()
```

# Matplotlib Subplot
**Display Multiple Plots**
- With the `subplot()` function we can draw multiple plots in one figure

```python
import matplotlib.pyplot as plt
import numpy as np
#pot 1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)
#pot 2
x=np.array([0,1,2,3])
y=np.array([8,9,10,11])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

# The subplot() function
- The `subplot()` function takes three arguments that describes the layout of the figure.
- The layout is oragnized in rows and columns,which are represented by the first and second argument.
- The third argument represented the index of the current plot.
- `plt.subplot(1,2,1)` the figure has 1 rows,2 columns and this plot is the first plot.


```python
#draw two plot of each other
import matplotlib.pyplot as plt
import numpy as np
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2,1,1)
plt.plot(x,y)
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([10,20,30,40])
plt.subplot(2,1,2)
plt.plot(x,y)
plt.show()

```

- We can draw as many plots we like on one figure.Just describe the number of rows,columns and the index of the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()
```

# Super Title
- We can add a title to the entire figure with the `suptitle()` function

```python
import numpy as np
import matplotlib.pyplot as plt
#Plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.plot(x,y)
plt.subplot(1,2,1)
plt.title('sales')

#plot2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.plot(x,y)
plt.subplot(1,2,2)
plt.title('income')
plt.suptitle('myshop')
plt.show()


```
