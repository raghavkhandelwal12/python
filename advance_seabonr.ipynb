# Basic plotting Functions
- Seaborn provides various basic plotting function such as **sns.scatterplot**,**sns.lineplot**,**sns.histplot**

# Pairplot
- A pairplot allows to visualize relationship between multiple paris of variable .It is useful for exploring multidimensional data.


```python
import seaborn as sns
#Apply the default theme
sns.set_theme()
#load an example dataset
tips=sns.load_dataset('tips')
#create a visualization
sns.relplot(
    data=tips,
    x='total_bill',y='tip',col='time',
    hue='smoker',style='smoker',size='size')
```

- **Hue**: use the `hue` parameter to distinguish between different categories.
- **Palette** : choose a color palette that is easy to distinguish.

# Facetgrid
- Facetgrid is used to plot multiple charts on the conditions of different categories in the dataset.
  

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')
#create a facetgrid
g=sns.FacetGrid(tips,col='time',row='smoker')
g.map(sns.histplot,'total_bill')
plt.show()
```

- **Column and Row**:Use meaningful variables for `col` and `rows` to compare different susbset.
- **Mapping**:Ensure the mapped function is appropriate for the data distribution.

# Heatmap
- Heatmaps are great for visualizing matrix-like data such as correlation matrices.

```python
#seaborn functions
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load example datasets
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
flights=sns.load_dataset('flights')
diamonds=sns.load_dataset('diamonds')
```

```python
# set the aesthetic style of the plots
sns.set(style='whitegrid')
```

```python
# scatter plot
#used to plot the relationship between two variable with a dot for each data point
sns.scatterplot(data=iris,x='sepal_length',y='sepal_width',hue='species')
plt.title('scatter plot of sepal length vs sepal width')
plt.show()
```

```python
#lineplot
# used to plot continuous data point as a line
sns.lineplot(data=flights,x='year',y='passengers')
plt.title('Line plot of passengers over years')
plt.show()
```

```python
#histogram
#used to plot the distribution of a single variabel.
sns.histplot(data=tips,x='total_bill',bins=30,kde=True)
plt.title('histogram of total bills')
plt.show()
```

```python
#boxplot
#use to show the distribution of a quantitative variable and to compare between levels of a categroical variable.
sns.boxplot(data=tips,x='day',y='total_bill',hue='day')
plt.title('Box plot of total bill by day')
plt.show()
```

```python
#violin plot
#used to show the distribution of the data across different categories.It combines aspects of a box plot and kernal density plot.
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True)
plt.title('violin plot of total bill by day and sex')
plt.show()
```

```python
#bar plot
#used to plot the mean value of a numeric variable for each category.
sns.barplot(data=tips,x='day',y='total_bill',errorbar=None)
plt.title('Bar Plot of Average Total Bill by Day')
plt.show()
```

```python
sns.barplot(data=tips,x='day',y='total_bill')
plt.title('Bar Plot of Average Total Bill by Day')
plt.show()
```

```python
#count plot
#used to show the counts of observatios in each categorical bin using bars.
sns.countplot(data=tips,x='day')
plt.title('count plot of days')
plt.show()
```

```python
#Heatmap
#used to display data in a matrix where values are represented with color intensity.
pivot_table=flights.pivot(index='month',columns='year',values='passengers')
sns.heatmap(pivot_table,annot=True,fmt='d',cmap='YlGnBu')
plt.title('heatmap of passengers')
plt.show()
```

```python
#Pair Plot
#used to plot pairwise relaionship in a dataset.It create a matrix of scatter plots
sns.pairplot(iris,hue='species')
plt.title('Pair plot of iris dataset')
plt.show()
```

```python
#joint plot
#used to draw a plot of two variable with bivariate and univariate graphs.
sns.jointplot(data=tips,x='total_bill',y='tip',kind='reg')
plt.title('joint plot of total bill and tip')
plt.show()
```
