# Different categories of plot in seaborn
- Plots are basically use for  visualizing the relationship between variables.Those variable can be either completely numerical or a catagory like a group,class,or division.Seaborn divides the plot into the below categories.

- **Relational Plots**: This plot is used to understand the relation between two variables.
- **Categorical plots**: This plot deals with categorical variable and how they can be visualized.
- **Distribution plots**: This plot is used for examining univariate and bivariate distributions.
- **Regressioon Plots**: The regression plots in seaborn are primarily intended to add a visual guide that helps to emphasize patterns in a dataset during exploratory data analyses.
- **Matrix Plots**: A matrix plot is an array of scatterplots.
- **Multi-plot grids**: It is a useful approach to draw multiple instances of the same plot on different subset of the dataset.

# Some basic plots using seaborn
**Histplot**: Seaborn Histplot used to visualize the univariate set of distribution(single variable).It plots a histogram with some other variations like kdeplot and rugplot.The Histplot function takes several arguments but the important onces are:

- **data**: This is the array,series or dataframe that we want to visulaize.It is required parameter.
- **x**: This specifies the column in the data to use for the histogram.If data is a dataframe we can specify the column by name.
- **y**: This specifies the column in the data to use for the histogram when you want to create a bivariate histogram. By default, it is set to None, meaning that a univariate histogram will be plotted.
- **bins** : This specifies the number of bins to use when dividing the data into intervals for plotting.By default it is set to auto which uses an algorithm to determine the optimal number of bins.
- **kde** : This parameter controls whether to display a kernal density estimate(kde) of the data in addition to the histogram .

```python
import numpy as np
import seaborn as sns
sns.set(style="white")

#genearte a random univariate dataset
rs=np.random.RandomState(10)
d=rs.normal(size=100)
#Plot a simple histogram and kde

sns.histplot(d,kde=True,color='m')
```

# Plotting graph using Seaborn|Python
- **Plotting categorical scatter plots with seaborn**
  

```python
# Python program to illustrate 
# Plotting categorical scatter 
# plots with Seaborn 

# importing the required module 
import matplotlib.pyplot as plt 
import seaborn as sns 

# x axis values 
x = ['sun', 'mon', 'fri', 'sat', 'tue', 'wed', 'thu'] 

# y axis values 
y = [5, 6.7, 4, 6, 2, 4.9, 1.8] 

# plotting strip plot with seaborn 
ax = sns.stripplot(x=x, y=y) 

# giving labels to x-axis and y-axis 
ax.set(xlabel='Days', ylabel='Amount_spend') 

# giving title to the plot 
plt.title('My first graph') 

# function to show plot 
plt.show() 
```

**Explanation**: This is the one kind of scatter plot of categorical data with the help of  seaborn.
- Categorical data is represented on the x-axis and values correspond to them represented through the y-axis.
- **.stripplot()** function is used to define the type of the plot and to plot them on canvas using.
- **.set()** function is used to set labels x-axis and y-axis
- **.tile()** function is used to give a title to the graph.
- To vies plot we use **.show() function**

# Barplot
- A **barplot** is basically used to aggregate the categorical data according to sme methods and by default its the mean.It can also be understood as a visualization of the group by action.To use this plot we choose a categorical columns for the x-axis and a numerical column for the y-axis and we see that it crates a plot taking a mean per categorical columns.
- **syntax**
- **barplot([x,y,hue,data,order,hue_order,])**

```python
import seaborn as sns
df=sns.load_dataset('tips')
#change the estimator from mean to 
#standard deviation
sns.barplot(x='sex',y='total_bill',data=df,palette='plasma')
```

**Explanation**:
- **Palette** is used to set the color of the plot.

# Boxplot
- Boxplot is a visula representing of the depicting groups of numerical data through their queatiles.Boxplot is also used to detect the outlier in the data set.

- **boxplot([x,y,hue,data,order,hue_order,..])**

**Grid type plot** : This example shows a regression plot of tips vs the total_bill from the dataset.**lmplot** stands for linear model plot and is used to create a regression plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
sns.lmplot(x ='total_bill',hue='total_bill', y ='tip', height = 2, aspect = 4, data = tips)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'tips' dataset
tips = sns.load_dataset('tips')

# Create a linear model plot with the correct parameters
sns.lmplot(x='total_bill', y='tip', height=2, aspect=4, data=tips)

# Display the plot
plt.show()

```

**Explanation**
- **sns.lmplot()**: Creates a linear model plot (regression plot) using the specified data. This plot shows the relationship between two variables along with a linear regression line.

- **x='total_bill**: Specifies that the total_bill variable from the tips DataFrame will be used for the x-axis. This variable represents the total bill amount for each dining occasion.
- **y='tip**: Specifies that the tip variable from the tips DataFrame will be used for the y-axis. This variable represents the tip amount given.
  
- **height=2**: Sets the height of each facet (individual plot) in inches. In this case, the height is set to 2 inches.
- **aspect=4**: Sets the aspect ratio of the plot, which is the ratio of the width to the height. An aspect ratio of 4 means the width is 4 times the height, making the plot long and narrow.
- **data=tips**: Specifies the DataFrame (tips) from which the data will be drawn.

# Scale and Context
- The set_context() allows us to override default parameter.This affect things like the size of the labels,lines and other element of the plot.
- The contex are

```python
#using poster
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('tips')
sns.set_context('poster',font_scale=0.01)
sns.countplot(x='day',hue='day',data=df,palette='coolwarm')
```

```python

#using notebook
import matplotlib.pyplot as plt
import seaborn as sns
df=sns.load_dataset('tips')
sns.set_context('notebook',font_scale=1)
sns.countplot(x='day',hue='day',data=df,palette='coolwarm')
```

# Seaborn.FacetGrid() method
- Seaborn is a python data visualization library used to matplotlib.It provides a high level interface for drawing attractive and informative statistical graphics.

- **seaborn.FacetGrid():**
- FacetGrid class helps in visualizing distribution of one variable as well as the relationship between multiple variable separately within subsets of our dataset using multiple panels.
- A FacetGrid can be drawn three dimensions?row,col and hue.The first two have obvious correspondance with the resulting array of axes think of the hue variable as a third dimesnion along a deph axis.
- FacetGrid object takes a dataframe as input and the names of the variables that will form the raow,column or hue dimensions of the grid.

```python
import seaborn
import matplotlib.pyplot as plt
df=seaborn.load_dataset('tips')
#form a facegrid using column with a hue
graph=seaborn.FacetGrid(df,col='sex',hue='day')
#map the above form facegrid with some attributes
graph.map(plt.scatter,'total_bill','tip',edgecolor='w').add_legend()
plt.show()
```

# How to make Histogram with Density Plots with Seaborn histplot?
- **Histogram** are visualizing tool that represent the distribution of a set of continuous data.In a histogram,the dta is dividid into a set of intervals or bins(usually on the x-axis) and the count of the data points that fall into each bins corresponding to the height of the bar above that bin.These bins may or may not be equal in width but are adjacent.
- A **density plot**(also known as kernal density plot) is another visualization tool for evaluating data distributions.It can be considered as a smoothed histogram.

- **Syntax: seaborn.histplot(data, x, y, hue, stat, bins, binwidth, discrete, kde, log_scale)**

# Python Seaborn-Catplot
- Seaborn helps resolve the two major problems faced by matplotlib:the problems are?
 - **Default Matplotlib parameter**
 - **working with data frames**
 - As seaborn compliments and extends Matplotlib,the learning curve is quite gradual.Seaborn library offers many advantages over other plotting libraries.
1. It is very easy to use and requires less code syntax
2. Works really well with pandas data structure,which is just what we needs as a data scientist.
3. It is built on top of Matplotlib,another vast and deep data visualization library.

- **syntax:**
- seaborn.catplot(*,x=None,y=None,hue=None,data=None,row=None,col=None,kind='strip',color=None,palette=None,**kwargs)
   

# Parameters
- **x,y,hue**:names of variable in data inputs for plotting long-form data.
- **data**:DataFrame Long-form(tidy) dataset for plotting.Each column should correspond to a variable ,and each row should be corrrespond to an observation.
- **row,col**:names of variable in data,optional categorical variable that will determine the faceting of the grid.
- **kind**:str,optimal the kind of plot to draw,corresponds to the name of a categorical axes-level potting function options are 'strip','swarm','box','violin','boxen','point','bar' or count
- **color**: Matplotlib color,optional color for all of the element or seed for a gradient palette.
- **palette**: name, list, or dict
Colors to use for the different levels of the hue variable. Should be something that can be interpreted by color_palette(), or a dictionary mapping hue levels to matplotlib colors.

```python

import seaborn as sns 
  
exercise = sns.load_dataset("exercise") 
g = sns.catplot(x="time", y="pulse", 
                hue="kind", 
                data=exercise)
```
